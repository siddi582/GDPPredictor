```python
# Import dependencies
import pandas as pd 
import numpy as np
```

```python
# Read CSV file
df = pd.read_csv("country_data.csv", encoding ="unicode_escape")
df.head()
```

```python
# Drop the series code and country code columns
df1 = df.drop(columns=['Series Code', 'Country Code'])
df1.head()
```

```python
# Drop country groupings such as "Africa Eastern and Southern", starting row 1521
df1=df1.iloc [:1520,:]
```

```python
# Rename the columns by assigning a new list of column names to the columns attribute. 
new_column_names = ['country_name', 'series_name', '1960','1961','1962','1963','1964','1965','1966','1967','1968','1969','1970','1971','1972','1973','1974','1975','1976','1977','1978','1979','1980','1981','1982','1983','1984','1985','1986','1987','1988','1989','1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']
df1.columns = new_column_names
df1.head(10)
```

```python
# Replace special characters ".." with naan value
df1=df1.replace("..","")
df1.replace('', np.nan, inplace=True)
```

```python
# Calculate the total number of missing values in the dataframe
missing_values = df1.shape[0] - df1.dropna().shape[0]
missing_values
```

```python
# Drop the years with more than 70% of missing data 

# Calculate the percentage of missing values in each column
missing_percent = df1.isnull().mean(axis=0)
print(missing_percent)

# Identify the columns that have more than 70% missing data
cols_to_drop = missing_percent[missing_percent > 0.70].index

# Drop the identified columns
df1 = df1.drop(cols_to_drop, axis=1)
df1.head()
```

```python
# Calculate the percentage of missing values in each row
missing_percent = df1.isnull().mean(axis=1)

# Identify the rows that have more than 70% missing data
rows_to_drop = missing_percent[missing_percent > 0.7].index

# Drop the identified rows
df1 = df1.drop(rows_to_drop)
```

```python
# Selecting the subset of the data that contains only the numeric values
data=df1.iloc[:,2:]

# Convert the subset dataframe values to float
#Creating a loop to convert the variables into float
for column in data:
    data[column] = data[column].astype(float)    

#Fill missing data with averages 
data = data.fillna(data.mean())
data
```

```python
#Verify there are no more rows with missing values
rows_with_missing_values = data.isna().any(axis=1)
rows_with_missing_values
```

```python
#Verify there are no more columns with missing values
data.isnull().all()
```

```python
# Create a non numeric dataframe containing country_name and series_name
non_num=df1.iloc[:,:2]

# Add back non numeric columns to the dataframe
df1 = pd.merge(non_num, data, left_index=True, right_index=True)
df1
```

```python
# Create the GDP per capita (current US$) dataframe by filtering
gdp_per_cap=df1[df1['series_name'] == 'GDP per capita (current US$)']
gdp_per_cap.head()
```

```python
# Turn dataframe from wide to long format
gdp_per_cap = gdp_per_cap.drop(columns=['series_name'])
gdp_per_cap = pd.melt(gdp_per_cap, id_vars =["country_name"], var_name = "Year", value_name = "gdp_per_cap" )
gdp_per_cap.head()
```

```python
# GDP growth (annual %)
gdp_growth=df1[df1['series_name'] == 'GDP growth (annual %)']
gdp_growth.head()
```

```python
# Turn dataframe from wide to long format
gdp_growth = gdp_growth.drop(columns=['series_name'])
gdp_growth = pd.melt(gdp_growth, id_vars =["country_name"], var_name = "Year", value_name = "gdp_growth" )
gdp_growth.head()
```

```python
# GDP (current US$)	dataframe
gdp_current=df1[df1['series_name'] == 'GDP (current US$)']
gdp_current.head()
```

```python
# Turn dataframe from wide to long format
gdp_current = gdp_current.drop(columns=['series_name'])
gdp_current = pd.melt(gdp_current, id_vars =["country_name"], var_name = "Year", value_name = "gdp_current" )
gdp_current.head()
```

```python
# Inflation, consumer prices (annual %) dataframe
inflation=df1[df1['series_name'] == 'Inflation, consumer prices (annual %)']
inflation.head()
```

```python
# Turn dataframe from wide to long format
inflation = inflation.drop(columns=['series_name'])
inflation = pd.melt(inflation, id_vars =["country_name"], var_name = "Year", value_name = "inflation" )
inflation.head()
```

```python
# Consumer price index (2010 = 100)	
cpi=df1[df1['series_name'] == 'Consumer price index (2010 = 100)']
cpi.head()
```

```python
# Turn dataframe from wide to long format
cpi = cpi.drop(columns=['series_name'])
cpi = pd.melt(cpi, id_vars =["country_name"], var_name = "Year", value_name = "cpi" )
cpi.head()
```

```python
# Unemployment, total (% of total labor force) (national estimate)
unemployment=df1[df1['series_name'] == 'Unemployment, total (% of total labor force) (national estimate)']
unemployment.head()
```

```python
# Turn dataframe from wide to long format
unemployment = unemployment.drop(columns=['series_name'])
unemployment = pd.melt(unemployment, id_vars =["country_name"], var_name = "Year", value_name = "unemployment" )
unemployment.head()
```

```python
# Current health expenditure per capita (current US$)
health_exp=df1[df1['series_name'] == 'Current health expenditure (% of GDP)']
health_exp.head()
```

```python
# Turn dataframe from wide to long format
health_exp = health_exp.drop(columns=['series_name'])
health_exp = pd.melt(health_exp, id_vars =["country_name"], var_name = "Year", value_name = "health_exp" )
health_exp.head()
```

```python
# Merge all dataframes 

merged_df1 = pd.merge(gdp_per_cap,gdp_current, right_index = False, left_index =False)
merged_df1.head()

merged_df2 = pd.merge(merged_df1,inflation, right_index = False, left_index =False)
merged_df2.head()

merged_df3 = pd.merge(merged_df2,unemployment, right_index = False, left_index =False)
merged_df2.head()

merged_df4 = pd.merge(merged_df3,cpi, right_index = False, left_index =False)
merged_df4.head()

merged_df5 = pd.merge(merged_df4,health_exp, right_index = False, left_index =False)
merged_df5.head()

merged_df6 = pd.merge(merged_df5,gdp_growth, right_index = False, left_index =False)
merged_df6.head()
```

```python
print(merged_df6.shape)
```

```python
# Remove outliers using IQR method
# Iterate through columns of the dataframe, starting from the 3rd column
for col in merged_df6.columns[2:]:
    # Calculate the first and third quartiles (Q1 and Q3)
    Q1 = merged_df6[col].quantile(0.25)
    Q3 = merged_df6[col].quantile(0.75)

    # Calculate the IQR
    IQR = Q3 - Q1

    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    # Identify the outlier countries
    outlier_countries = merged_df6[(merged_df6[col] < lower_bound) | (merged_df6[col] > upper_bound)]

    # Remove the outlier countries from the dataset
    merged_df6 = merged_df6[~merged_df6.index.isin(outlier_countries.index)]

print(merged_df6.shape)
```
